# Data cleaning

```python
print(f"Duplicates: \n\
Tracks- {tracks.duplicated().sum()}\n\
Description- {desc.duplicated().sum()}")
```

```python
tracks = tracks.drop_duplicates()
```

```python
# Data type conversion
tracks["duration_ms"] = tracks["duration_ms"].astype("int64")
tracks["key"] = tracks["key"].astype("int64")
tracks["mode"] = tracks["mode"].astype("int64")
tracks["time_signature"] = tracks["time_signature"].astype("int64")
tracks["language"] = tracks["language"].astype("category")
```

```python
tracks.info()
```

```python
# Decoding key, mode, time_signature
# as they are numerical in storage, categorical in meaning

key_map = {0:"C", 1:"C#", 2:"D", 3:"D#", 4:"E", 5:"F", 6:"F#", 7:"G", 8:"G#",
           9:"A", 10:"A#", 11:"B"}
tracks['key_cat'] = tracks['key'].map(key_map) # keeping original column same

mode_map = {0:'Minor', 1:'Major'}
tracks['mode_cat'] = tracks['mode'].map(mode_map)

tracks['time_signature_cat'] = tracks['time_signature'].astype(str)+"/4" # eg., 4/4
```

```python
tracks.info()
```

```python
# Checking for outliers

def check_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Column: {col}")
    print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)\n")
    return outliers


pop_outliers = check_outliers(tracks, "popularity")
dur_outliers = check_outliers(tracks, "duration_ms")
dan_outliers = check_outliers(tracks, "danceability")
ene_outliers = check_outliers(tracks, "energy")
```

```python
# Removing outliers in popularity, danceability, duration_ms, energy

def clean_tracks(col):
  Q1 = tracks[col].quantile(0.25)
  Q3 = tracks[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  tracks[(tracks[col] >= lower_bound) & (tracks[col] <= upper_bound)]

clean_tracks("popularity")
clean_tracks("duration_ms")
clean_tracks("danceability")
clean_tracks("energy")
```

5. **Recent Dip (Post-2020)**
